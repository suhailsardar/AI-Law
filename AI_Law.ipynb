```python
import streamlit as st
from transformers import AutoTokenizer, AutoModelForCausalLM

# Define the model name
model_name = "AdaptLLM/law-chat"

# Load the tokenizer and model
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name)

# Function to generate text from the model
def generate_text(input_text, max_length=5000, temperature=0.7, top_k=50, top_p=0.9):
    # Encode the input text
    input_ids = tokenizer.encode(input_text, return_tensors='pt')

    # Generate text
    output = model.generate(
        input_ids,
        max_length=max_length,
        temperature=temperature,
        top_k=top_k,
        top_p=top_p,
        num_return_sequences=1
    )

    # Decode the output
    decoded_output = tokenizer.decode(output[0], skip_special_tokens=True)

    return decoded_output

# Streamlit UI
st.title("Law Chatbot")

# Input text from user
input_text = st.text_area("Enter your text:", "Explain the main principles of intellectual property law.")

# Generate and display output when the button is pressed
if st.button("Generate"):
    with st.spinner("Generating response..."):
        generated_text = generate_text(input_text)
        st.write("**Generated Text:**")
        st.write(generated_text)

```
